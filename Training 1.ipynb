```python
X=house_data.drop(['price','id','date','long','view','waterfront'],axis=1)
y=house_data['price']
```

```python
X.head()
```

```python
from sklearn.feature_selection import SelectKBest, chi2,f_regression
```

```python
selector=SelectKBest(chi2,k=14)
selector.fit_transform(X,y)
selector.get_support()
```

```python
X=X[np.array(X.columns[selector.get_support()])]
```

```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import SGDClassifier,Ridge,Lasso
from sklearn.preprocessing import StandardScaler,PolynomialFeatures,RobustScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVC,LinearSVC
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
model=make_pipeline(SimpleImputer(strategy='median'),RobustScaler(),Lasso(alpha=10,max_iter=100000))
#model=make_pipeline(SimpleImputer(),StandardScaler(),Lasso(alpha=0.2,max_iter=10000))
#model=Ridge(alpha=0.01)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
params={'fit_intercept':['True','False'],'alpha':[0.8,1,2,3,4,5],'selection':['cyclic', 'random'],'max_iter':[1000,5000,10000,100000]}
grid=GridSearchCV(model,param_grid=params, cv=4)

grid.fit(X_train,y_train)
```
