```python
df =df.select_dtypes(include=['int','float'])

headers=list(df.columns.values)
fields=[]

for field in headers:
    fields.append( {
        'name': field,
        'mean':df[field].mean(),
        'var':df[field].var(),
        'sdev':df[field].std()
    }
    )
for field in fields:
    print(field)
```

```python
newDf=pd.DataFrame(fields)
display(newDf)
```

```python
newDf.drop('name',axis=1,inplace=True) # axis 0 for rows and axis 1 for columns
```
