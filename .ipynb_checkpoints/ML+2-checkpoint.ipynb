```python
# importer les données avec pandas et prendre le necéssaire
# Il faut découper les données avec train test 
#ensuite faire un choix de modèle
# faire le gridSearch pour trouver les meilleurs paramètres du modèle
# faire une validation set pour éviter l'overfit par exemple
# Utiliser les learning curve pour estimer le nombre de données nécessaires pour améliorer le modèle
```

```python
import sklearn
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# exemple importation model
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
X=iris.data
y=iris.target
print(X.shape)
plt.scatter(X[:,0],X[:,1],c=y,alpha=0.8)
plt.show()
```

```python
# diviser data set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
model=KNeighborsClassifier()
```

```python
model.fit(X_train,y_train)
print("Train score:{}".format(model.score(X_train,y_train)))
print("Test score:{}".format(model.score(X_test,y_test)))
```

```python
# Validation set
#on cherche le modèle avec le meilleur validation set
# Cross vaélidation pour valider sur plusieurs coupes
from sklearn.model_selection import cross_val_score,KFold,LeaveOneOut,ShuffleSplit,StratifiedKFold, GroupKFold
# Kfold
    #cv=KFold(5,random_state=0)
#Cas particulier Kfold LeaveOneOut
    #cv=LeaveOneOut()
# ShuffleSplit
    #cv=ShuffleSplit(5,test_size=0.2) # nombre de fois qu'on fait le test, pourcentage de test
# StratifiedKfold (le bon compromis)
cv=StratifiedKFold(5)
# GroupKfold
cv=GroupKFold(5).get_n_splits(X,y,groups=X[:,0])
cross_val_score(model,X_train,y_train,cv=cv,scoring='accuracy').mean()
```

```python
#validation curve pour choisir le bon paramètre à régler
# exemple de k ici
from sklearn.model_selection import validation_curve
k=np.arange(1,50)
train_score,val_score=validation_curve(model,X_train,y_train,'n_neighbors',k,cv=5)
plt.plot(k,val_score.mean(axis=1),label='validation')
plt.plot(k,train_score.mean(axis=1),label='Train')
plt.ylabel('score')
plt.xlabel('n_neighbors')
plt.legend()
plt.show()
```

```python
# Test de toutes les combinaisons de paramètres d'un modèle
from sklearn.model_selection import GridSearchCV
param_grid={'n_neighbors':np.arange(1,20),'metric':['euclidean','manhattan']}
grid=GridSearchCV(model,param_grid,cv=3)
grid.fit(X_train,y_train)
# modèle avec le meilleur score
print(grid.best_score_)
print(grid.best_params_)
```

```python
Bmodel=grid.best_estimator_
```

```python
Bmodel.score(X_test,y_test)
```

```python
# confusion matrix pour comprendre les erreurs
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,model.predict(X_test))
```

```python
# Les perfomances augmentent t-elles si on donne plus de données? avec le meilleur modèle évidemment
from sklearn.model_selection import learning_curve
N,train_score,val_score=learning_curve(Bmodel,X_train,y_train,train_sizes=np.linspace(0.1,1.0,20),cv=5)
print(N)
plt.plot(N,train_score.mean(axis=1),label='train')
plt.plot(N,val_score.mean(axis=1),label='validation')
plt.xlabel('train sizes')
plt.legend()
plt.show()
```
